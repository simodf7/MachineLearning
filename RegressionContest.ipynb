### Rimozione delle righe Outlier con il metodo Z-score

```python
import pandas as pd
import numpy as np
from scipy import stats

# Function to remove outliers using z-scores calculated from the training data
def removing_outliers(train_data, train_target):
    z_scores = np.abs(stats.zscore(train_data))
    mask = (z_scores < 3).all(axis=1)
    train_data_filtered = train_data[mask]
    train_target_filtered = train_target[mask]
    return train_data_filtered, train_target_filtered
```

### Gestione Data Leakage

Il dataset contiene il risultato degli esami di vari soggetti. Risulta pertanto fondamentale non andare incontro al fenomeno di Data Leakage. 
Esso si verifica quando lo splitting fa sì da invalidare l'indipendenza dei due set di Training e Test. 
A tal scopo, andiamo a estrapolare l'ID del soggetto dall'ID associato all'istanza.


```python
training_data['Subject'] = training_data['ID'].str.extract(r'sub-(\d{4})')
training_data = training_data.drop(columns = ['ID'])
training_data
```

```python
from sklearn.svm import SVR
import numpy as np

param_grid = {
    'C': np.linspace(20,30),
    'epsilon': np.linspace(0.1, 1, 4)
}

model = SVR(kernel ='rbf')
```

### Funzione per la rimozione delle feature correlate 

Sarà utilizzata insieme alla pipeline di trasformazioni definita nelle celle successive. Andiamo quindi a calcolare la correlazione tra le varie feature e andiamo a rimuovere quelle che mostrano alta correlazione. 

```python
# Rimozione delle feature altamente correlate
def remove_highly_correlated_features(train_data, test_data, threshold=0.8):
    corr_matrix = pd.DataFrame(train_data).corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    train_data_reduced = pd.DataFrame(train_data).drop(columns=to_drop).values
    test_data_reduced = pd.DataFrame(test_data).drop(columns=to_drop).values
    return train_data_reduced, test_data_reduced
```

### Nested Cross Validation 


![title](PseudoCode.png)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GroupKFold
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import VarianceThreshold
from itertools import product


X = training_data.drop(columns=['Age', 'Subject'])
Y = training_data['Age']
groups = training_data['Subject']


pipeline = make_pipeline(
    StandardScaler(),
    VarianceThreshold(0.01)
)

outer_cv = GroupKFold(n_splits=10)

mae_test = []

for train_index, test_index in outer_cv.split(X, Y, groups):
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]   


    groups_for_inner = groups.iloc[train_index]
    
    inner_cv = GroupKFold(n_splits=5)
    validation_params = []

    i=1
    keys, values = zip(*param_grid.items())
    for v in product(*values):
        print(i)
        i += 1
        params = dict(zip(keys, v))
        model.set_params(**params)
        mae_scores = []
        
        for train_val_index, val_index in inner_cv.split(X_train, y_train, groups_for_inner):
            X_train_val, X_val = X_train.iloc[train_val_index], X_train.iloc[val_index]
            y_train_val, y_val = y_train.iloc[train_val_index], y_train.iloc[val_index]
    
            X_train_val = pipeline.fit_transform(X_train_val)
            X_val = pipeline.transform(X_val)
    
            X_train_val, X_val = remove_highly_correlated_features(X_train_val, X_val)
    
            pca = PCA(n_components = 0.95)
            X_train_val_transform = pca.fit_transform(X_train_val)
            X_val_transform = pca.transform(X_val)       
            
            
            model.fit(X_train_val_transform, y_train_val)
            y_pred_val = model.predict(X_val_transform)
            mae_single = mean_absolute_error(y_val, y_pred_val)
            mae_scores.append(mae_single)

        
        validation_params.append({'params': params, 'avg_score': sum(mae_scores)/5})
    

    best_avg_score_valid = float('inf')
    best_avg_params_valid = None

    for entry in validation_params:
        avg_score = entry['avg_score'] 
        if avg_score < best_avg_score_valid:
            best_avg_score_valid = avg_score
            best_avg_params_valid = entry['params']
    
    print("Best parameters based on average score:", best_avg_params_valid)
    print("Average score:", best_avg_score_valid)
    

    X_train = pipeline.fit_transform(X_train)
    X_test = pipeline.transform(X_test)

    X_train, X_test = remove_highly_correlated_features(X_train, X_test)

    pca = PCA(n_components = 0.95)
    X_train_transform = pca.fit_transform(X_train)
    X_test_transform = pca.transform(X_test)    
    
    model.set_params(**best_avg_params_valid)
    model.fit(X_train_transform, y_train)

    y_pred = model.predict(X_test_transform)

    mae = mean_absolute_error(y_test, y_pred)
    print("MAE on test set:", mae)
    mae_test.append(mae)


print("External CV Test Error:", sum(mae_test)/10)
```

### Definizione del modello finale

Procediamo con la procedura per settare il miglior modello. 
Splittiamo in Training e Test sempre rispettando i gruppi. Eseguiamo una CV sul training per ogni combinazione degli iperparametri per stimare al meglio i migliori iperparametri. Trovata la migliore combinazione addestriamo sull'intero training e valutiamo sul test. 

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GroupKFold, GroupShuffleSplit
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import VarianceThreshold
from itertools import product


X = training_data.drop(columns=['Age', 'Subject'])
y = training_data['Age']
groups = training_data['Subject']


pipeline = make_pipeline(
    StandardScaler(),
    VarianceThreshold(0.01)
)


train_inds, test_inds = next(GroupShuffleSplit(random_state = 42).split(X, y, groups))
X_train, X_test, y_train, y_test = X.iloc[train_inds], X.iloc[test_inds], y.iloc[train_inds], y.iloc[test_inds]

groups_for_val = groups.iloc[train_inds]

validation_params = []

i=0
keys, values = zip(*param_grid.items())
for v in product(*values):
    i += 1
    print(i)
    params = dict(zip(keys, v))
    model.set_params(**params)
    mae_param = []
    
    for train_val_index, val_index in GroupKFold().split(X_train, y_train, groups_for_val):
            X_train_val, X_val = X_train.iloc[train_val_index], X_train.iloc[val_index]
            y_train_val, y_val = y_train.iloc[train_val_index], y_train.iloc[val_index]

            X_train_val, y_train_val = removing_outliers(X_train_val, y_train_val)
        
            X_train_val = pipeline.fit_transform(X_train_val)
            X_val = pipeline.transform(X_val)
    
            X_train_val, X_val = remove_highly_correlated_features(X_train_val, X_val)
    
            pca = PCA(n_components = 0.95)
            X_train_val_transform = pca.fit_transform(X_train_val)
            X_val_transform = pca.transform(X_val)        
    
            model.fit(X_train_val_transform, y_train_val)
            y_pred_val = model.predict(X_val_transform)
            mae_single = mean_absolute_error(y_val, y_pred_val)
            mae_param.append(mae_single)
        
    
    validation_params.append({'params': params, 'avg_score': sum(mae_param)/5})
    

# Troviamo la migliore combinazione di iperparametri all'interno della lista
best_avg_score_valid = float('inf')
best_avg_params_valid = None

for entry in validation_params:
    avg_score = entry['avg_score']
    if avg_score < best_avg_score_valid:
        best_avg_score_valid = avg_score
        best_avg_params_valid = entry['params']

print("Best parameters based on average score:", best_avg_params_valid)
print("Average score:", best_avg_score_valid)

X_train, y_train = removing_outliers(X_train, y_train)

X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)


X_train, X_test = remove_highly_correlated_features(X_train, X_test)
    
pca = PCA(n_components = 0.95)
X_train_transform = pca.fit_transform(X_train)
X_test_transform = pca.transform(X_test)        

model.fit(X_train_transform, y_train)
y_pred_test = model.predict(X_test_transform)
mae = mean_absolute_error(y_test, y_pred_test)
print("Mae on Test Set:", mae)
```
